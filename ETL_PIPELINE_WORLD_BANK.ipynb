## Instead of converting the xml file into a dataframe, we can simply iterate through the file and display its content

```python
def print_lines(n, file):
    '''
    n: number of lines being read
    file:name of the file
    '''
    f = open(file)
    for i in range(n):
        print(f.readline())
    f.close()
```

```python
print_lines(30, 'population_data.xml')
```
